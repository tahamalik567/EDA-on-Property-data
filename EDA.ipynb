```python
import pandas as pd
```

```python
# Load the Excel file into a DataFrame
file_path = "Data.xlsx"
df = pd.read_excel("Data.xlsx")
```

```python
# Display the first few rows of the DataFrame to understand its structure
print(df.head())
```

```python
print(df.info())
```

```python
# Summary statistics for numerical columns
print(df.describe())
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include='number')

# Display the first few rows of the numeric DataFrame
print(numeric_df.head())
```

![alt text](correlation-1536x778.webp)

```python
# Correlation matrix to identify relationships between numerical columns
correlation_matrix = numeric_df.corr()
print(correlation_matrix)
```

```python
# Visualize correlation matrix (optional but helpful)
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 9))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
